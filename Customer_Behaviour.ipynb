```python
# ==========================================
# Customer Shopping Behavior Analysis
# ==========================================

import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus
from IPython.display import display

# Load Dataset
df = pd.read_csv("customer_shopping_behavior.csv")

print("========== FIRST 5 ROWS ==========")
display(df.head())

print("\n========== DATASET INFORMATION ==========")
df.info()

print("\n========== SUMMARY STATISTICS ==========")
display(df.describe())

print("\n========== MISSING VALUES ==========")
display(df.isnull().sum())

print("\n========== FILLING MISSING VALUES ==========")

df["Review Rating"] = (
    df.groupby("Category")["Review Rating"]
      .transform(lambda x: x.fillna(x.median()))
)

print("Remaining Missing Values:")
display(df["Review Rating"].isnull().sum())

print("\n========== STANDARDIZING COLUMN NAMES ==========")

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.rename(columns={"purchase_amount_(usd)": "purchase_amount"}, inplace=True)

display(df.columns)

print("\n========== CREATING AGE GROUPS ==========")

labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]

df["age_group"] = pd.qcut(
    df["age"],
    q=4,
    labels=labels
)

display(df[["age", "age_group"]].head())

print("\n========== PURCHASE FREQUENCY MAPPING ==========")

frequency_mapping = {
    "Weekly": 52,
    "Bi-Weekly": 26,
    "Fortnightly": 26,
    "Monthly": 12,
    "Quarterly": 4,
    "Every 3 Months": 4,
    "Annually": 1
}

df["frequency_per_year"] = df["frequency_of_purchases"].map(frequency_mapping)

display(df[["frequency_of_purchases", "frequency_per_year"]].head())

print("\n========== CONNECTING TO POSTGRESQL ==========")

username = "postgres"
password = quote_plus("Tanu@1107")
host = "localhost"
port = "5432"
database = "customer_behaviour"

engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

print("Database Connected Successfully!")

print("\n========== EXPORTING DATA TO POSTGRESQL ==========")

df.to_sql(
    "customer_behaviour",
    engine,
    if_exists="replace",
    index=False
)

print("Data Exported Successfully!")

print("\n========== FINAL DATASET ==========")
display(df.head())
```
